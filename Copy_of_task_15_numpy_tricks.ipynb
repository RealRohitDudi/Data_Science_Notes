###`Q-1:` Create a random 3x4 matrix with value between 0-100. And perform below tasks

    i. Sort this matrix. np.sort()
    ii. Sort this matrix based on values in 2nd column.
    iii. Sort this matrix based on max value in each row.
    iv. Sort based on elements value.


See examples:
```
arr =
    [[92 90 74]
    [ 6 63 93]
    [15 93 96]
    [70 60 48]]
```
```
i. np.sort
[[74 90 92]
 [ 6 63 93]
 [15 93 96]
 [48 60 70]]

ii. based on 2nd column
[[70 60 48]
 [ 6 63 93]
 [92 90 74]
 [15 93 96]]

iii. based on row max- ascending
[[15 93 96]
 [ 6 63 93]
 [92 90 74]
 [70 60 48]]

iv. based on elements value
[[ 6 15 48]
 [60 63 70]
 [74 90 92]
 [93 93 96]]

```

````python
# code here
# i. Sort this matrix. np.sort()
# ii. Sort this matrix based on values in 2nd column.
# iii. Sort this matrix based on max value in each row.
# iv. Sort based on elements value.
import numpy as np
arr = np.random.randint(1,100,12).reshape(3,4)
print(arr)
print("sorted: \n",np.sort(arr))
arr[:,1] = np.sort(arr[:,1],axis=-1)
print("2nd column sorted: \n", arr)

for row in arr:
    row.sort()
print("max value in each row: \n", arr)

flat_sort = np.sort(arr,axis=None)
print("based on elements value: \n", flat_sort)
````

###`Q-2:` There is an array of marks of 5 students in 4 subjects. Further you are asked to perform below task.
    i. Add marks every student of an extra subject in the same array.
    ii. Add two new students marks in respective 5 subjects.(one subject added in above task)
    iii. Add extra column with sum of all subjects(5-subjects) marks
    iv. Sort the array(non-ascending order) on total marks column--one added in above task. Show top 2 rows.

Note: Change dimension of arrays during concatenation or appending if required.

Given Array-
```
marks = [[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]]

extra_subject = [41, 87, 72, 36, 92]
#Two extra students record-
rec1 = [77, 83, 98, 95, 89]
rec2 = [92, 71, 52, 61, 53]
```

````python
# code here
# students  = row
marks = np.array([[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]])

extra_subject = np.array([41, 87, 72, 36, 92]).reshape(-1,1)
merged = np.hstack((marks,extra_subject))
print("marks after adding extra subject: \n", merged)
#do not touch til

student1 = np.array([31, 85, 75, 66, 52]).reshape(1,-1)
student2 = np.array([31, 27, 73, 46, 32]).reshape(1,-1)


updated_students = np.vstack((merged,student1,student2))
print("after adding 2 more students \n",updated_students)


total_marks = np.sum(updated_students,axis=1).reshape(-1,1)
sum_added = np.hstack((updated_students,total_marks))
print("after adding sum column",sum_added)


sum_added[:,5] = np.sort(sum_added[:,5])[::-1]
print("after sorting based on sum column descening()",sum_added)

````

### `Q-7:` You are given a array. You have to find the minimum and maximum array element and remove that from the array.

```python
import numpy as np

np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))
```

````python
# code here
arr = np.random.randint(100, 1000, 200).reshape((1, 200))

print("original array: \n", arr)
print("maximum array \n", arr.min())
print("minimum array \n",arr.max())

min_index = np.argmin(arr)
max_index = np.argmax(arr)

arr = np.delete(arr,min_index)
arr = np.delete(arr,max_index)
print("after deleting min and max: \n", arr)
````

### `Q-9:` You are given a array ($arr ∈ [0, 1]$). First you have round off the elements upto 3 decimal places and compare that
- 0th percentile == minimum value of the array
- 100th percentile == maximum value of the array
- also find the difference betwen 51th percenile and 50th percentile values

````python
# code here
arr = np.random.random(100)
print(arr)
rounded_arr = np.round(arr,decimals=3)
print("rounded arr: \n",rounded_arr)

zero_th_percentile = np.percentile(rounded_arr,0)
hundred_th_percentile = np.percentile(rounded_arr,100)
print("zero_th_percentile",zero_th_percentile)
print("hundred_th_percentile",hundred_th_percentile)

print("is min == 0th percentile: ",rounded_arr.min()==zero_th_percentile)
print("is max == 100th percentile: ",rounded_arr.max()==hundred_th_percentile)
````
