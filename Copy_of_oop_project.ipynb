```python
import nlpcloud

class NLPApp:

  def __init__(self):
    self.__database = {}
    self.__menu1()


  def __menu1(self):
    input1 = input("""Enter 1 for registration:
    Enter 2 for login:
    Enter 3 for exit: """)

    if input1 == '1':
      self.__register()
    elif input1 == '2':
      self.__login()
    else:
      exit()


  def __register(self):
    name = input('enter name')
    email = input('enter email')
    password = input('enter password')
    if email in self.__database:
      print('email already exists')
    else:
      self.__database[email] = [name,password]
      print('Registration successful.')
      print(self.__database)
      self.__login()

  def __login(self):
    email = input('Enter email: ')
    password = input('Enter password: ')
    if email in self.__database:
      if self.__database[email][1] == password:
        print("Login successful")
        self.__menu2()
      else:
        print('Wrong password. Try again')
        self.__login()
    else:
      print('This email is not registered')


  def __menu2(self):
    input2 = input("""What do you want to use?
    Enter 1 for NER:
    Enter 2 for emotion analysis:
    Enter 3 for generate Headline:
    Enter 4 for logout:
    Enter anything for main menu """)

    if input2 == '1':
      self.__ner()
    elif input2 == '2':
      self.__emotion_analysis()
    elif input2 == '3':
      self.__headline_generator()
    else:
      self.__menu1()


  def __emotion_analysis():
    print("Start emotion analysis in text...")
    para = input('Enter the paragraph: ')
    client = nlpcloud.Client("gpt-oss-120b", "32e0404d189937a81ee71b3818b1cd6a8fe1f4ca", gpu=True)
    response = client.sentiment(para,  target="NLP Cloud")
    print("Result: ", response)

  def __ner():
    print("Start Entity extraction in text...")
    para = input('Enter the paragraph: ')
    entity = input('Entity to search: ')
    client = nlpcloud.Client("gpt-oss-120b", "32e0404d189937a81ee71b3818b1cd6a8fe1f4ca", gpu=True)
    response = client.entities(para, searched_entity=entity)
    print("Result: ", response)


  def __headline_generator():
    print("Start headline generator...")
    para = input('Enter text to generate headline: ')
    client = nlpcloud.Client("t5-base-en-generate-headline", "32e0404d189937a81ee71b3818b1cd6a8fe1f4ca", gpu=False)
    response = client.summarization(para)
    print("Result: ", response)


app1 = NLPApp()
```
